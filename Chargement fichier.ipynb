```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
import requests
 
# URL directe vers le fichier Excel
url = "https://www.nyc.gov/assets/nypd/downloads/excel/crime_statistics/park-crime/nyc-park-crime-stats-q3-2024.xlsx"
file_path = "/lakehouse/default/Files/nyc-park-crime-stats-q3-2024.xlsx"
 
# Télécharger le fichier dans le Lakehouse
response = requests.get(url)
with open(file_path, "wb") as f:
    f.write(response.content)
 
print(f"Fichier téléchargé avec succès dans : {file_path}")
```
